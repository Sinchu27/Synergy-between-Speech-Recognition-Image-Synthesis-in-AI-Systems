```python
import pyttsx3 as p  # Importing the text-to-speech conversion library
import speech_recognition as sr  # Importing the speech recognition library

engine = p.init()  # Initializing the text-to-speech engine
rate = engine.getProperty('rate')  # Getting the current speech rate
engine.setProperty('rate', 180)  # Setting the speech rate to 180 words per minute
voices = engine.getProperty('voices')  # Getting the list of available voices
engine.setProperty('voice', voices[0].id)  # Setting the voice to the second voice in the list

def speak(text):
    engine.say(text)  # Passing the text to the engine to be spoken
    engine.runAndWait()  # Running the speech engine to speak the text

r = sr.Recognizer()  # Initializing the recognizer for speech recognition
speak("Hello. I am your voice assistant. How are you?")  # Greeting the user

with sr.Microphone() as source:  # Using the microphone as the audio source
    r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
    r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
    print("Listening....")  # Indicating that the program is listening
    audio = r.listen(source)  # Listening to the audio input
    try:
        text = r.recognize_google(audio)  # Recognizing the speech using Google's API
        print(text)  # Printing the recognized text
        speak("I am also having a good day ma'am")  # Responding to the user
    except sr.UnknownValueError:
        print("Sorry, I did not understand that.")  # Handling unrecognized speech
        speak("Sorry, I did not understand that.")  # Informing the user about the issue

def perform_task():
    speak("What can I do for you?")  # Asking the user what task to perform

    with sr.Microphone() as source:  # Using the microphone as the audio source
        r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
        r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
        print("Listening....")  # Indicating that the program is listening
        audio = r.listen(source)  # Listening to the audio input
        try:
            text2 = r.recognize_google(audio).lower()  # Recognizing and converting the speech to lowercase text
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")  # Handling unrecognized speech
            speak("Sorry, I did not understand that.")  # Informing the user about the issue
            return

    if "information" in text2:
        speak("What is the topic information that you need?")  # Asking for the topic information
        with sr.Microphone() as source:
            r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
            r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
            print("Listening....")  # Indicating that the program is listening
            audio = r.listen(source)  # Listening to the audio input
            try:
                infor = r.recognize_google(audio)  # Recognizing the speech
                speak("Getting the information about {}".format(infor))  # Responding to the user
            except sr.UnknownValueError:
                print("Sorry, I did not understand that.")  # Handling unrecognized speech
                speak("Sorry, I did not understand that.")  # Informing the user about the issue
                return
        assist = wiki()  # Assuming wiki is a class for fetching Wikipedia information
        assist.get_info(infor)  # Fetching information from Wikipedia
        
    elif "play" in text2 and "video" in text2:
        speak("Which video do you want me to play?")  # Asking for the video to play
        with sr.Microphone() as source:           
            r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
            r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
            print("Listening....")  # Indicating that the program is listening
            audio = r.listen(source)  # Listening to the audio input
            try:
                vid = r.recognize_google(audio)  # Recognizing the speech
                print("Playing {} on YouTube".format(vid))  # Printing the video to be played
                speak("Playing {} on YouTube".format(vid))  # Responding to the user
            except sr.UnknownValueError:
                print("Sorry, I did not understand that.")  # Handling unrecognized speech
                speak("Sorry, I did not understand that.")  # Informing the user about the issue
                return
        assist = yt()  # Assuming yt is a class for playing YouTube videos
        assist.get_info(vid)  # Playing the specified video on YouTube
        
    elif "generate" in text2 and "image" in text2:
        speak("Which image do you want me to generate?")  # Asking for the image to generate
        with sr.Microphone() as source:
            r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
            r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
            print("Listening....")  # Indicating that the program is listening
            audio = r.listen(source)  # Listening to the audio input
            try:
                gen = r.recognize_google(audio)  # Recognizing the speech
                print("Generating the image of {}".format(gen))  # Printing the image to be generated
                speak("Generating the image of {}".format(gen))  # Responding to the user
            except sr.UnknownValueError:
                print("Sorry, I did not understand that.")  # Handling unrecognized speech
                speak("Sorry, I did not understand that.")  # Informing the user about the issue
                return
        assist = img()  # Assuming img is a class for generating images
        assist.get_info(gen)  # Generating the specified image

while True:
    perform_task()  # Performing the specified task
    print("Do you want me to perform another task?")  # Asking if another task is to be performed
    speak("Do you want me to perform another task?")  # Asking if another task is to be performed
    
    with sr.Microphone() as source:
        r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
        r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
        print("Listening....")  # Indicating that the program is listening
        audio = r.listen(source)  # Listening to the audio input
        try:
            text3 = r.recognize_google(audio).lower()  # Recognizing and converting the speech to lowercase text
        except sr.UnknownValueError:
            print("Sorry, I did not understand that.")  # Handling unrecognized speech
            speak("Sorry, I did not understand that.")  # Informing the user about the issue
            continue  # Continuing the loop to ask again
        
    if "yes" in text3:
        print("Which one do you want me to perform?")  # Asking which task to perform
        speak("Which one do you want me to perform?")  # Asking which task to perform
        
        with sr.Microphone() as source:
            r.energy_threshold = 10000  # Setting the energy threshold for ambient noise
            r.adjust_for_ambient_noise(source, 1.2)  # Adjusting for ambient noise
            print("Listening....")  # Indicating that the program is listening
            audio = r.listen(source)  # Listening to the audio input
            try:
                text4 = r.recognize_google(audio).lower()  # Recognizing and converting the speech to lowercase text
                if "generate image" in text4 or "play a video" in text4 or "information" in text4:
                    continue  # Continuing to perform the requested task
            except sr.UnknownValueError:
                print("Sorry, I did not understand that.")  # Handling unrecognized speech
                speak("Sorry, I did not understand that.")  # Informing the user about the issue
                continue  # Continuing the loop to ask again
                
    elif "no" in text3:
        print("Thank you. Signing off. Have a good day!")  # Printing a sign-off message
        speak("Thank you. Signing off. Have a good day!")  # Speaking the sign-off message
        break  # Breaking the loop to end the program
```
